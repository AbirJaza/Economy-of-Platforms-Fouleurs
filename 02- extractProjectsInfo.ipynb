```python
l = []
for link in ProjectsLinks:
    ProjectName = link.split('/')[2]
    driver.get(fouleursURL+link)
    time.sleep(0.3)
    Information = driver.find_element_by_xpath("//div[@class='cooked']")
    print('STEP : {}'.format(link))
    l.append((ProjectName,Information.text.replace('\\n','; ').replace('\n','; ').replace('\\r', '').replace('\\',' ')))
```
